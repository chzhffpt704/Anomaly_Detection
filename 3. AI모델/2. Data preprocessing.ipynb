```python
import sys
import pandas as pd
import numpy as np
import os
sys.path.append('D:/이근섭/88. 취미 및 다양한 활동/1. python 관련 연습/1. 기계시설물 고장 예지 센서/')
import Feature_extractor as fe
```

```python
def Data_load(CSV_path):
    df = pd.read_csv(CSV_path, skiprows= 9, usecols= [0,1], names = ['time', 'acceleration'])

    Hz = pd.read_csv(CSV_path, skiprows= 4, nrows= 1, header= None) #모터 속도, 정격출력(watt), 정격전류(amp)
    Sample_rate = pd.read_csv(CSV_path, skiprows= 6, nrows= 1, header= None, usecols= [1])

    N = df.shape[0] # data size
    Fs = Sample_rate[1][0] # 샘플링 주파수 -> ex) 1000 Hz : 초당 1000개의 샘플이 있다
    return (df, Fs, Hz[2][0], Hz[3][0], Hz[4][0])
```

```python
%%time
import time
nor_path = "../1. Dataset/"
#Train Val
Train_Val_folder = os.listdir(nor_path)
Train_Val_folder = [Train_Val_folder[0]]

for Train_Val in Train_Val_folder:
    Feature_list = []
    Train_Val_path = os.path.join(nor_path,Train_Val) 
    kw_folder = os.listdir(Train_Val_path)
    for kw in kw_folder:
        kw_path = os.path.join(Train_Val_path, kw)
        Device_folder = os.listdir(kw_path)
        for Device in Device_folder:
            Device_path = os.path.join(kw_path, Device)
            label_folder = os.listdir(Device_path)
            start_time = time.time()
            for label in label_folder:
                label_path = os.path.join(Device_path, label)
                data_folder = os.listdir(label_path)
                
                print(Train_Val, kw, Device, label, len(data_folder))
                
                for data in data_folder:
                    data_path = os.path.join(label_path, data)
                    # Data load
                    df, Fs, Hz, watt, amp = Data_load(data_path)
                    # Feature extract
                    Fe_ex = fe.Feature_extractor(df)
                    Freq_ex = fe.Freq_extractor(df, Hz, Fs)
                    AbsMax, AbsMean, P2P, RMS, Skewness, Kurtosis, CrestFactor, ShapeFactor, ImpulseFactor =  Fe_ex.Feature()
                    f_1x, f_2x, f_3x, f_4x =  Freq_ex.Freq_Freatures()
                    
                    Feature_tmp = [kw, Device, label,
                        Fs, Hz, watt, amp,
                    AbsMax, AbsMean, P2P, RMS, Skewness, Kurtosis, CrestFactor, ShapeFactor, ImpulseFactor,
                    f_1x, f_2x, f_3x, f_4x]
                    Feature_list.append(Feature_tmp)
                end_time = time.time() - start_time
                print(f"소요 시간 {end_time}")
    # Data Save            
    columns = ["kw","Device","label",
        "Fs", "Hz", "watt", "amp", "AbsMax", "AbsMean", "P2P", "RMS", 
        "Skewness", "Kurtosis", "CrestFactor", "ShapeFactor", "ImpulseFactor", 
        "f_1x", "f_2x", "f_3x", "f_4x"
    ]
    Feature_df = pd.DataFrame(Feature_list, columns= columns)
    Feature_df.to_csv(f"{Train_Val.split()[1]}.csv",index = False)
    





    
```
